```python
import json
```

```python
def load_data(filename):
    with open( filename ,"r") as f:
        data = json.load(f)

    return data    
```

```python
data = load_data("store_data.json")
print(data)
```

```python
# clean the data
def clean_data(data):
    for user in data:
        print(user)
```

```python
clean_data(data)
```

```python
# clean ratings}

text_to_num = {"one" : 1 , "two" :2 , "three" :3 , "four" : 4 , "five" :5}
for user in data:
    raw_rating = user["rating"].strip().lower()
    if(raw_rating in text_to_num):
        raw_rating = text_to_num[raw_rating]
    user["rating"] = raw_rating
    
```

```python
# handle missing values
raw_age = user.get("age")
if(raw_age == None):
    user['age'] = None
```

```python
cleaned_data = []
unique_users = set()   

for user in data:      
    name = user["name"].strip()
    
    if name in unique_users:
        continue
    
    unique_users.add(name)
    cleaned_data.append(user)

print(cleaned_data)

```

```python
def get_insights(cleaned_data):
    total_rating = 0
    for user in cleaned_data:
        total_rating += float(user["rating"])
    print(f"avg rating = {total_rating/len(cleaned_data)}")

    poor_rating = 0;
    for user in cleaned_data:
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"poor rating percentage is {poor_rating/len(cleaned_data)*100}")
```

```python
get_insights(cleaned_data)
```

```python
def recommendations(cleaned_data):
    recommendations = []


    for user in cleaned_data:
        curr_recomm ={}
        curr_recomm["name"] = user["name"]


        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "samsung"


        recommendations.append(curr_recomm)
    return recommendations
```

```python
recommendations(cleaned_data)
```
