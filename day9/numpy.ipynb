```python
arr11 = np.array([[1,2] , [3 , 4]])
mean = np.mean(arr11)
std_dev = np.std(arr11)

normalised_array = (arr11-mean)/std_dev
print(normalised_array)
print(mean)
print(std_dev)
```

```python
arr12 = np.array([1,2,4,5,6,7,8])
print(np.sum(arr12))
print(np.product(arr12))
print(np.min(arr12))
print(np.max(arr12))
print(np.argmin(arr12))
print(np.argmax(arr12))
print(np.mean(arr12))
print(np.std(arr12))
print(np.median(arr12))
print(np.var(arr12))
```

```python
arr13 = np.array([1,2,3,5,6,7,8,9,10,0])
print(np.square(arr13))
print(np.sqrt(arr13))
print(np.log(arr13))
print(np.log2(arr13))
print(np.exp(arr13))
print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.sort(arr13))
```
